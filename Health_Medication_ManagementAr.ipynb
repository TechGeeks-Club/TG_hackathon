# تطبيق تتبع الصحة مع إدارة الأدوية

</div>


<div dir="rtl" align="right">

## 1-وصف المشكلة:

بصفتنا مالكي الشركة الطبية "MMT"، نسعى إلى إنشاء تطبيق أو موقع إلكتروني حصريًا لعملائنا لمساعدتهم على إدارة صحتهم بشكل فعال. يحتاج عملاؤنا إلى أداة تمكنهم من مراقبة نشاطهم البدني، نظامهم الغذائي، نومهم، والالتزام بتناول أدويتهم، مع السماح لأطبائنا بمتابعة تقدمهم عن بُعد. حاليًا، يواجه عملاؤنا صعوبة في البقاء منظمين ومتحفزين في رحلتهم الصحية، كما أن أطبائنا يفتقرون إلى نظام مركزي لمراقبة البيانات في الوقت الفعلي.

</div>


<div dir="rtl" align="right">

سيسمح هذا التطبيق لعملائنا بـ:

 - تتبع أنشطتهم اليومية (التمارين الرياضية، النظام الغذائي، النوم).

 - إدارة أدويتهم من خلال التذكيرات وسجلات التتبع-. 

 - مشاركة حالتهم الصحية مع أطبائنا للحصول على متابعة شخصية.

 
في الوقت نفسه، سيوفر لأطبائنا أداة لـ: 


مراقبة تقدم مرضاهم عن بُعد  .

- تلقي تنبيهات في حالة حدوث مشكلات (مثل تفويت الأدوية أو وجود حالة صحية مقلقة).


</div>


<div dir="rtl" align="right">

## 2-الفئة المستهدفة:
**- الأفراد المهتمون بالصحة:** لتتبع نشاطهم البدني، نظامهم الغذائي، نومهم، وأدويتهم.

**- المرضى الذين يتلقون علاجًا طبيًا:** لإدارة أدويتهم ومشاركة حالتهم الصحية مع طبيبهم.

**- الأطباء والمهنيون الصحيون:** لمتابعة تقدم مرضاهم عن بُعد.


</div>

<div dir="rtl" align="right">

## 3-الأهداف:
مساعدة المستخدمين على تتبع نشاطهم البدني، نظامهم الغذائي، نومهم، وأدويتهم.

تمكين المرضى من مشاركة حالتهم الصحية مع طبيبهم.

تقديم توصيات مخصصة لتحسين صحتهم ورفاهيتهم.

تحفيز المستخدمين من خلال التذكيرات، الإشعارات، ومخططات التقدم.

</div>

<div dir="rtl" align="right">

## 3-الميزات المدمجة:
**- لوحة التحكم: تصور البيانات الصحية الرئيسية.

**- دفتر الأدوية:** تتبع الأدوية مع تسجيل التاريخ، الوقت، ووصف الحالة الصحية.

**- التذكيرات:** إشعارات للأدوية والأنشطة اليومية.

**- واجهة الطبيب:** منصة للأطباء لمتابعة المرضى عن بُعد.


</div>

<div dir="rtl" align="right">

## 4-المخرجات:


نموذج تفاعلي على Figma (نماذج قابلة للنقر).

تطبيق ويب كامل الوظائف (واجهة أمامية وخلفية).

واجهة مستخدم intuitiva واستجابة.

نظام إشعارات فعال.

لوحة تحكم للأطباء تتضمن أدوات تحليل ومراقبة المرضى.

عرض تقديمي (خطابي) لعرض فكرة التطبيق ووظائفه.
